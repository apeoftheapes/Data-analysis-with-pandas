```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, date


df = pd.read_csv('BATS_AMZN, 15.csv')
df['time'] = df['time'].apply(lambda x: datetime.fromtimestamp(x))

df = df[(df['time'] >= datetime.fromisoformat('2022-01-03')) & (df['time'] < datetime.fromisoformat('2022-01-05')) ]

filt = df[(df['time'] >= datetime.fromisoformat('2022-01-03')) & (df['time'] < datetime.fromisoformat('2022-01-05')) ]

first_period = df[(df['time'].dt.hour < 16)]
print(first_period)
first_period['day'] = first_period['time'].dt.year * 365 + first_period['time'].dt.month * 31 + first_period['time'].dt.day
first_period_high = first_period.groupby('day')['time', 'high'].max()
first_period_low = first_period.groupby('day')['time', 'low'].min()

print(first_period_high)
print(first_period_low)

df['day'] = df['time'].dt.year * 365 + df['time'].dt.month * 31 + df['time'].dt.day


fig, ax1 = plt.subplots(figsize=(10,4))
x = df['time']
y1 = df['close']

ax1.plot(x, y1, '-', color = (208/255, 208/255, 208/255))

ax1.step(first_period_high['time'], first_period_high['high'], '-', color='g', where='post')
ax1.step(first_period_low['time'], first_period_low['low'], '-', color='r', where='post')

ax1.set_xlabel('Datetime')
ax1.xaxis_date()

ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d- %H:%M'))
fig.autofmt_xdate()

ax1.set_ylabel('price')

# 145.9455  146.9495  145.6905  146.9375
```
